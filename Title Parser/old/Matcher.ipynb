```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import os

df_collected=pd.DataFrame()
# Loop until the sample size is less than or equal to the total number of titles
for x in range(df_post.shape[0]+1):
    df_post_sample = df_post[x:x+1]
    df_manual_titles_sample = df_manual_titles

    # Vectorize the sampled titles
    vectorizer = TfidfVectorizer().fit_transform(df_post_sample['title'].tolist() + df_manual_titles_sample['title'].tolist())

    # Compute cosine similarity
    cosine_sim = cosine_similarity(vectorizer)

    # Get the number of rows in df_post_sample
    num_rows = df_post_sample.shape[0]

    # Get the cosine similarity scores for pairs where one element is from df_post_sample and the other is from df_manual_titles_sample
    sim_scores = cosine_sim[:num_rows, num_rows:]

    # Set a threshold for similarity. This can be adjusted based on your specific requirements
    threshold = 0.4

    # Get a boolean mask where the similarity score is above the threshold
    mask = (sim_scores.max(axis=1) > threshold)

    # Apply the mask to df_post to get the rows where the title is similar to a title in df_manual_titles
    df_post_filtered = df_post_sample[mask]


    # Append df_post_filtered to the bottom of df
    df_collected = pd.concat([df_collected, df_post_filtered], ignore_index=True)
    # Save df_post_filtered as a CSV file
    #df_post_filtered.to_csv('df_post_filtered.csv', index=False, mode='a', header=not os.path.exists('df_post_filtered.csv'))
```

```python
df_collected.to_csv('df_post_filtered.csv', index=False)
```
