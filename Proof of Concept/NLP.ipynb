```python
import pandas as pd

df_significant_words = pd.read_csv('significant_words.csv')
```

```python
df_significant_words
```

```python
import matplotlib.pyplot as plt

top_20_words = df_significant_words.head(20)
plt.figure(figsize=(12, 6))
plt.bar(top_20_words['Word'], top_20_words['Frequency'])
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Top 30 Words')
plt.xticks(rotation=45, ha='right')
plt.show()
```
